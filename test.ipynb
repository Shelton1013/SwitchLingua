```python
import requests
import datetime
import time
import jsonlines as jsonl

# Constants
API_KEY = "241af5c234c39e440673b659eece63c5"  # Replace with your own API key
BASE_URL = "https://gnews.io/api/v4/top-headlines"
COUNTRY = "hk"
CATEGORY = "general"
OUTPUT_FILE = "news_data.json"

# Initialize storage for articles
all_articles = []

# Get today's date in UTC
today = datetime.datetime.utcnow()

# Loop through the last 100 days
title_hash = {}
for i in range(100):
    # Calculate the target date
    target_date = today - datetime.timedelta(days=i)
    
    # Format the date to the required format (ISO 8601 with 'Z' for UTC)
    from_date = target_date.strftime("%Y-%m-%dT00:00:00Z")
    to_date = target_date.strftime("%Y-%m-%dT23:59:59Z")

    # Construct API request
    params = {
        "country": COUNTRY,
        "category": CATEGORY,
        "apikey": API_KEY,
        "from": from_date,
        "to": to_date,
    }

    try:
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status()  # Raise an error for HTTP issues
        data = response.json()

        if "articles" in data:
            articles = data["articles"]
            for article in articles:
                if article['title'] in title_hash:
                    continue
                else:
                    with jsonl.open("news/news_data.jsonl", "a") as f:
                        f.write(article)
            print(f"Fetched {len(articles)} articles for {target_date.date()}")
        else:
            print(f"No articles found for {target_date.date()}")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for {target_date.date()}: {e}")
    
    time.sleep(1)
    # Respect API rate limits # Pause for 1 second between requests


print(f"Finished fetching. {len(all_articles)} articles saved to {OUTPUT_FILE}")
```
